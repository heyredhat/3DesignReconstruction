```python
n, r, alpha = 3, 3, 8
P = construct_P(n, r, alpha)
print(P)
P_data = extract_P_data(P)
Phi = P_data["Phi"]
valid, ratio = valid_probabilities(1000, P)
print(np.all(valid @ Phi @ valid.T>=0))
viz(P, m=5000, embed=True)
```

```python
n, r, alpha = 3, 3, 2
P = construct_P(n, r, alpha)
print(P)
A = rand_symmetric_tensor(n, 3)
print(valid_probabilities(1, P, A=A))
C = C_constructor(P, A=A)
print([np.all(np.linalg.eigvals(C(p)) >= 0) for p in P])
viz(P, A=A, embed=True)
```
